```python
user_book_df = book_df.pivot_table(index='User_ID',
                                 columns='Book_Title',
                                 values='Book_Rating')
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
user_book_df.index = book_df.User_ID.unique()
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values, metric='cosine')
```

```python
user_sim
```

```python
np.fill_diagonal(user_sim, 0)
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book_df.User_ID.unique()
user_sim_df.columns = book_df.User_ID.unique()
```

```python
#Considering the most similar customer
def recommend(custID):
    simID = user_sim_df.loc[custID].idxmax()
    simID_book = book_df[book_df['User_ID'] == simID].Book_Title
    custID_book = book_df[book_df['User_ID'] == custID].Book_Title
    return set(simID_book) - set(custID_book)
```

```python
#Considering top 5 similar customers
def recommend_new(custID):
    simID = list(user_sim_df.sort_values(custID,ascending = False).head(5).index)
    simID_book = [book_df[book_df['User_ID'] == each].Book_Title for each in simID]
    simID_book = np.concatenate(simID_book)
    custID_book = book_df[book_df['User_ID'] == custID].Book_Title
    return set(simID_book) - set(custID_book)
```

```python
recommend_new(162121)
```

```python
recommend_new(276726)
```
